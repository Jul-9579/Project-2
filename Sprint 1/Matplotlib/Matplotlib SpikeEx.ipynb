```python
# Functional Method

plt.plot(x,y)

plt.show()
```

```python
# add labels and title
plt.plot(x,y)

plt.xlabel('X Label here')
plt.ylabel('Y Label here')
plt.title('Title')
plt.show()
```

```python
# Object oriented method

fig = plt.figure() # imagine a blank canvas

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('1st plot ')
```

```python
axes  # its an array
```

```python
# meaning we can iterate through it
fig, axes = plt.subplots(nrows= 1, ncols = 3)

for i in axes :
    i.plot(x,y)
```

```python
# since we can iterate, this means we can also go through the indexes

fig, axes = plt.subplots(nrows= 1, ncols = 2)

axes[0].plot(x,y)
axes[0].set_title('1st Plot')

axes[1].plot(y,x)
axes[1].set_title('2nd Plot')




plt.tight_layout()
```

```python
# legend

# Figure size

fig = plt.figure(figsize=(6,4))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label = 'X squared')
ax.plot(x,x**3, label = 'X cubed')
plt.xlabel("hello")
ax.legend()
```

```python
# legend

# Figure size

fig = plt.figure(figsize=(6,4))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x*2, label = 'X squared')
ax.plot(x,x*3, label = 'X cubed')


ax.legend(loc = 1)
```

```python
# setting colors

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y , color = '#FF8C00')  # google RGB HEX Code #FF8C00
```

```python
# linewidth

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y , color = 'purple', lw = 50)  # googlg RGB HEX Code #FF8C00
```

```python

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y , color = 'purple', linewidth = 3, alpha = 0.5 ) # transparency . @ lw=  instead of linewidth
```

```python
# linestyle


fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y , color = 'purple', linewidth = 3, alpha = 0.5 , linestyle = 'Steps') # ls= '-.' 'steps' , ';'
```

```python


fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y , color = 'purple', linewidth = 3, alpha = 0.5 , linestyle = '--' , marker = 'o') # x, o , * , +
```

```python

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y , color = 'purple', linewidth = 3, alpha = 0.5 , linestyle = '--' , marker = 'o', markersize = 20 , markerfacecolor = 'yellow',
        markeredgecolor = 'green', markeredgewidth = 3 ) # x, o , * , +
```

```python
# limit X , Y label

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y , color = 'purple', linewidth = 3, alpha = 0.5 , linestyle = '--'  ) # x, o , * ,

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
